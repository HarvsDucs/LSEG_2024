```python
import pandas as pd

gdp = pd.read_csv("C:/Users/Harvey/Downloads/economic_data_nov17/WORLD_monthly_gdp.csv")
US_unemployment = pd.read_csv("C:/Users/Harvey/Downloads/economic_data_nov17/US_monthly_unemployment_rate.csv")
US_inflation = pd.read_csv("C:/Users/Harvey/Downloads/economic_data_nov17/US_monthly_inflation.csv")

display(gdp.head(2))
print("tail")
display(gdp.tail(2))
display(US_unemployment.head(2))
print("tail")
display(US_unemployment.tail(2))
display(US_inflation.head(2))
print("tail")
display(US_inflation.tail(2))
```

```python
# Reversing economic dataframes so that new dates will appear at top

US_unemployment = US_unemployment.iloc[::-1].reset_index(drop=True)
US_inflation = US_inflation.iloc[::-1].reset_index(drop=True)

display(US_unemployment.head(2))
print("tail")
display(US_unemployment.tail(2))
display(US_inflation.head(2))
print("tail")
display(US_inflation.tail(2))
```

```python
# Getting just the first 51 rows of the unemployment and inflation data

US_unemployment = US_unemployment[:51]
US_inflation = US_inflation[:51]
```

```python
# Reversing back

US_unemployment = US_unemployment.iloc[::-1].reset_index(drop=True)
US_inflation = US_inflation.iloc[::-1].reset_index(drop=True)

display(US_unemployment.head(2))
print("tail")
display(US_unemployment.tail(2))
display(US_inflation.head(2))
print("tail")
display(US_inflation.tail(2))
```

```python
# gdp["gdp_close"] = gdp["close"]
US_unemployment["US_unemployment_close"] = US_unemployment["close"]
US_inflation["US_inflation_close"] = US_inflation["close"]

US_unemployment.drop("close", axis='columns',inplace=True)
US_unemployment.drop("Volume", axis='columns',inplace=True)

US_inflation.drop("time", axis='columns',inplace=True)
US_inflation.drop("close", axis='columns',inplace=True)
US_inflation.drop("Volume", axis='columns',inplace=True)
```

```python
# Merging all dataframes
AI_index = pd.read_csv("C:/Users/Harvey/Downloads/AI_index_nov17/AI_index.csv")

merged_df = pd.concat([US_unemployment, US_inflation], axis=1)
merged_df_final = pd.concat([merged_df, AI_index], axis=1)

display(merged_df_final)
```

```python
import matplotlib.pylab as plt

X = merged_df_final[['US_unemployment_close', 'US_inflation_close']].values
y = merged_df_final['Index'] 

# plt.scatter(X[:, 0], X[:, 1], c='blue') just testing this out but isn't correct
```

```python
# Train test splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline 

pipe = Pipeline([
    ("scale", StandardScaler()),
    ("model", LinearRegression())
])

pred = pipe.fit(X, y).predict(X)
plt.scatter(pred, y)

```
